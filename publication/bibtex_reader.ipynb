```python
# https://stackoverflow.com/questions/9235853/convert-bibtex-file-to-database-entries-using-python
import bibtexparser
import pandas as pd

with open("publication_bib2.txt") as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)
    
df = pd.DataFrame(bib_database.entries)
# selection = df[['doi', 'number']]
df.to_csv('temp.csv', index=False)
```

```python
'''sort pandas dataframe'''
# https://stackoverflow.com/questions/37787698/how-to-sort-pandas-dataframe-from-one-column
# Use sort_values to sort the df by a specific column's values:
# df.sort_values('year')
df = df.sort_values(by=['year'], ascending=False)
df.head(5)
```

```python
'''iterate dataframe'''
print('\\begin{enumerate}')
# https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas
for index, row in df.iterrows():
    '''string.format python'''
    pub = "\\item \\publication[{0}]".format(row['ID'])
    '''string replace python'''
#     https://www.w3schools.com/python/ref_string_replace.asp
    pub = pub.replace("[", "{")
    pub = pub.replace("]", "}")
    print('\t' +pub)
#     print(row['ID'], row['year'])
print('\\end{enumerate}')
```
