```python
import requests
from bs4 import BeautifulSoup
import json

def extract_numista_features(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        features = {}
        
        # Numista organizes features in a table with id 'coin_features'
        features_table = soup.find('table', {'id': 'coin_features'})
        
        if features_table:
            rows = features_table.find_all('tr')
            for row in rows:
                header = row.find('th')
                data = row.find('td')
                
                if header and data:
                    key = header.get_text(strip=True)
                    # Clean up the value (remove extra whitespace/newlines)
                    value = " ".join(data.get_text().split())
                    features[key] = value
        
        return features

    except Exception as e:
        return {"error": str(e)}

# URL for the 50 Yen - Shōwa (1967-1988 version)
url = "https://en.numista.com/catalogue/pieces1026.html"
coin_data = extract_numista_features(url)

# Output as JSON
print(json.dumps(coin_data, indent=4, ensure_ascii=False))
```

```python
import requests
from bs4 import BeautifulSoup
import json

def scrape_numista(url):
    print(f"Connecting to: {url}...")
    
    # Headers are crucial to avoid being blocked
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept-Language': 'en-US,en;q=0.9',
    }
    
    try:
        session = requests.Session()
        response = session.get(url, headers=headers, timeout=10)
        
        # Check if the request was successful
        if response.status_code != 200:
            return {"error": f"Failed to load page. Status code: {response.status_code}"}

        soup = BeautifulSoup(response.text, 'html.parser')
        features = {}

        # Look for the specific table by ID
        features_table = soup.find('table', {'id': 'coin_features'})
        
        if not features_table:
            # Fallback: Look for any table that contains the word "Issuer" 
            # (Common in the features section)
            features_table = soup.find('th', string='Issuer').find_parent('table') if soup.find('th', string='Issuer') else None

        if features_table:
            for row in features_table.find_all('tr'):
                th = row.find('th')
                td = row.find('td')
                if th and td:
                    key = th.get_text(strip=True)
                    value = td.get_text(strip=True).replace('\xa0', ' ')
                    features[key] = value
            return features
        else:
            return {"error": "Could not find the Features section on this page."}

    except Exception as e:
        return {"error": f"An error occurred: {str(e)}"}

# Target URL (The specific 1967-1988 Shōwa 50 Yen)
coin_url = "https://en.numista.com/catalogue/pieces1026.html"

data = scrape_numista(coin_url)

# Print the result
if "error" in data:
    print(f"❌ {data['error']}")
else:
    print("✅ Success! Data extracted:")
    print(json.dumps(data, indent=4, ensure_ascii=False))
```
